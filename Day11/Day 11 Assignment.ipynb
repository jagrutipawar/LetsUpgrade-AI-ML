Now we calculate correlation for attrition with other clm.

```python
dataset.Attrition[dataset.Attrition == 'Yes'] = 1
dataset.Attrition[dataset.Attrition == 'No'] = 0
```

```python
dataset.head()
```

```python
from scipy.stats import pearsonr
# Attrition with age:
stats,p=pearsonr(dataset.Attrition,dataset.Age)
print("stats=",stats, "p=",p)
if p<0.05:
    print("variable +vely correlated")
elif p>0.05:
    print("variable -vely correlated")
else:
    print("Variable is stable")
```
