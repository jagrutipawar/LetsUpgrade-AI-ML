# Mannwhitney Test:

```python
import pandas as pd

dataset_yes= pd.read_excel('general_data.xlsx',sheet_name="Attrition_Yes")
dataset_no= pd.read_excel('general_data.xlsx',sheet_name="Attrition_No")

from scipy.stats import mannwhitneyu

stats,p = mannwhitneyu(dataset_yes.Age, dataset_no.Age)

print("Stats=",stats,"And P=",p)
if p<0.05:
    print("Hypothesis is rejectected.")
else:
    print("Alternate hypothesis accepted")
```
